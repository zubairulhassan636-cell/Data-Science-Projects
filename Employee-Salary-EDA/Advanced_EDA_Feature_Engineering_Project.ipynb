```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
from scipy.stats import zscore

df=pd.read_csv('Employee_Salary_Dataset.csv')
df.head()
```

## Dataset Overview

```
print(df.shape)
print(df.info())
display(df.describe(include='all'))
display(df.isnull().sum())
```

## Missing Value Handling

```
num=df.select_dtypes(include=np.number).columns
df_mean=df.copy()
for c in num:
    df_mean[c]=df_mean[c].fillna(df_mean[c].mean())
imputer=KNNImputer(n_neighbors=3)
df_knn=df.copy()
df_knn[num]=imputer.fit_transform(df_knn[num])
df=df_mean.copy()
```

## Outlier Detection & Treatment (IQR)

```
for c in num:
    Q1=df[c].quantile(.25);Q3=df[c].quantile(.75);IQR=Q3-Q1
    low=Q1-1.5*IQR;up=Q3+1.5*IQR
    df[c]=df[c].clip(low,up)
```

## Feature Engineering

```
if {'Salary','Experience_Years'}.issubset(df.columns):
    df['Salary_per_Year']=df['Salary']/(df['Experience_Years'].replace(0,np.nan))
if 'Age' in df.columns:
    df['Age_Group']=pd.cut(df['Age'],bins=[0,30,45,100],labels=['Young','Adult','Senior'])
if 'Experience_Years' in df.columns:
    df['Experience_Level']=pd.cut(df['Experience_Years'],bins=[-1,2,5,100],labels=['Beginner','Intermediate','Expert'])
df.head()
```

## EDA Visualizations

```
df.hist(figsize=(10,6))
plt.show()
for c in num:
    plt.figure(figsize=(5,3))
    plt.boxplot(df[c].dropna(),vert=False)
    plt.title(c)
    plt.show()
if len(num)>=2:
    sns.heatmap(df[num].corr(),annot=True,cmap='Blues')
    plt.show()
```

## Conclusion
- Missing values handled.
- Outliers treated using IQR.
- Three new features engineered.
- Dataset is cleaned and ready for machine learning.
